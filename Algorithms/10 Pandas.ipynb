Create a pie chart presenting the percentage of travelers categorised based on adulthood status and gender, before and after the unfortunate voyage. Given one's gender and age, abstracting any additional information, calculate their survival probability.

```python
beg_cat_stat=[males_under18.shape[0],males_over18.shape[0],females_under18.shape[0],females_over18.shape[0]]
beg_cat_stat#number of travelers based on adulthood status and gender [boys,men,girls,women] at the start
```

```python
beg_cat_stat=np.array(beg_cat_stat)
beg_cat_stat
```

```python
males_under18_sur = males_under18[(males_under18['Survived'] ==1)]
males_over18_sur = males_over18[(males_over18['Survived'] ==1)]
females_under18_sur = females_under18[(females_under18['Survived'] ==1)]
females_over18_sur = females_over18[(females_over18['Survived'] ==1)]
males_under18_sur
```

```python
end_cat_stat=[males_under18_sur.shape[0],males_over18_sur.shape[0],females_under18_sur.shape[0],females_over18_sur.shape[0]]
end_cat_stat=np.array(end_cat_stat)
end_cat_stat#number of travelers based on adulthood status and gender [boys,men,girls,women] at the start
```
