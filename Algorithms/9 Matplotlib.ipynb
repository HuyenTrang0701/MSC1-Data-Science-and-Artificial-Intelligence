```python
# Create Figure and Subplots
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,4), sharey=True, dpi=120)
# Plot
ax1.plot([1,2,3,4,5], [1,2,3,4,10], 'go')
ax2.plot([1,2,3,4,5], [2,3,4,5,21], 'b*')

# Title, X and Y labels, X and Y Lim
ax1.set_title('Scatterplot Greendots')
ax2.set_title('Scatterplot Bluestars')

#Axis labels
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax2.set_xlabel('X')
ax2.set_ylabel('Y') 

#ax limits
ax1.set_xlim(0, 6)  
ax2.set_xlim(0, 6)


ax1.set_ylim(0, 12) 
ax2.set_ylim(0, 12)

plt.tight_layout()
plt.show()
```

Plot a sine wave, from 0 to 10, and cosine wave from 2, 8. Color the sine wave blue, while the cosine one orange. Place labels what each curve represents. 

```python
plt.plot(np.linspace(0,10,100),np.sin(np.linspace(0,10,100)), c='blue', label='sine wave')
plt.plot(np.linspace(2,8,100),np.cos(np.linspace(2,8,100)), c='orange', label='cosine wave')
plt.legend()
```

Scatter points whose first cordinate follows the values of the sine wave above, while the second cordinate follows those of the cosine wave above.

```python
plt.scatter(np.sin(np.linspace(0,10,100)),np.cos(np.linspace(2,8,100)))
```

Create a loop with 10 steps such that np.cos(np.linspace(2,8,100)) slowly becomes np.cos(np.linspace(0,10,100)), and plot the results, as above.

```python
for i in range(1,11):
    plt.scatter(np.sin(np.linspace(0,10,100)),np.cos(np.linspace(2-i*0.2,8+i*0.2,100)))
    plt.show()
    plt.clf()
```

Create a 3D plot for the function f(x,y): sin(x)+cos(y), such that the boundaries of the plot are -10 and 10 in each axis. 

```python
def f(x, y):
    return np.cos(x)+np.sin(y)

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure(figsize=(15,10))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 500, cmap='autumn')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
```
