**Monte Carlo Simulation - Double Coin Flip**

A person will flip two fair coins. Use a Monte Carlo simulation in order to approximate the probability of getting two heads, 1 head and 1 tail, or 2 tails.

```python
list1=[0]*3
for i in range(100000):
    r1=random.randint(0,1)
    r2=random.randint(0,1)
    if r1==0 and r2==0:
        list1[0]+=1
    elif r1==1 and r2==1:
        list1[2]+=1
    else:
        list1[1]+=1
list1
```

```python
list(map(lambda x: x/sum(list1),list1))
```

**Create a Monte Carlo Simulation of the Monty Hall Problem. There are three doors. Behind one of the doors there is a car. While behind the other two there is a goat. The player chooses a door first. Then the moderator of the game opens one of the doors that has a goat behind. Two doors remain, the door the player chose and the door that was not opened by the moderator. The player is given an option to change his decision. Use Monte Carlo to find out whether the player should!**

```python
import random
won_switch=0
for i in range(100):
    door_car=random.randint(0,2)
    list1=[0]*3
    list1[door_car]=1
    print(list1)
    list2=list1[1:]
    list2.remove(0)
    list1=list1[:1]+list2
    print(list1)
    if list1[1]==1:
        won_switch+=1
```

```python
won_switch
```

```python
import random
won_idle=0
for i in range(10000):
    door_car=random.randint(0,2)
    list1=[0]*3
    list1[door_car]=1
    print(list1)
    list2=list1[1:]
    list2.remove(0)
    list1=list1[:1]+list2
    print(list1)
    if list1[0]==1:
        won_idle+=1
```

```python
won_idle
```

```python
[won_switch/(won_switch+won_idle),won_idle/(won_switch+won_idle)]
```

**Use a Monte Carlo Simulation to determine which is the tower Gandalf should visit first today, in order to maximize the chances of finding Saruman. Which is the second best choice?**

```python
states_0=[500,300,200]
states_1=[0,0,0]
```

```python
import random
for i in range(states_0[0]):
    r=random.uniform(0, 1)
    if r<0.2:
        states_1[0]+=1
    elif r<0.7:
        states_1[1]+=1
    else:
        states_1[2]+=1

for i in range(states_0[1]):
    r=random.uniform(0, 1)
    if r<0.4:
        states_1[0]+=1
    elif r<0.5:
        states_1[1]+=1
    else:
        states_1[2]+=1
        
for i in range(states_0[2]):
    r=random.uniform(0, 1)
    if r<0.3:
        states_1[0]+=1
    elif r<0.5:
        states_1[1]+=1
    else:
        states_1[2]+=1
```

```python
cmlt_prob_matrix=[[0.2,0.7],[0.4,0.5],[0.3,0.5]]
```

```python
states_through_time=[[5000,3000,2000]]
```

```python
iter_nr=50                                       #for 50 days
for t in range(iter_nr):
    states_through_time.append([0,0,0])
    for j in range(len(states_through_time[0])):
        for i in range(states_through_time[t][j]):
            r=random.uniform(0, 1)
            if r<cmlt_prob_matrix[j][0]:
                states_through_time[t+1][0]+=1
            elif r<cmlt_prob_matrix[j][1]:
                states_through_time[t+1][1]+=1
            else:
                states_through_time[t+1][2]+=1        
```
